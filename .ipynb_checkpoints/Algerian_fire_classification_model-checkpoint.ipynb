 ## KNeighborsClassifier

```python
from sklearn.pipeline import Pipeline

#setup the pipeline steps:steps
steps=[('scaler',StandardScaler()),('knn',KNeighborsClassifier())]

#create the pipeline
pipeline = Pipeline(steps)

# Create train and test sets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)

# Fit the pipeline to the training set: knn_scaled
knn_scaled = pipeline.fit(X_train,y_train)

# Instantiate and fit a k-NN classifier to the unscaled data
knn_unscaled = knn.fit(X_train,y_train)

# Compute and print metrics
print('Accuracy with Scaling: {}'.format(knn_scaled.score(X_test,y_test)))
print('Accuracy without Scaling: {}'.format(knn_unscaled.score(X_test,y_test)))
```

## Hyperparameter Tuning

```python
def Classifier_exercise(clf,Xtrain,Xtest,Ytrain,Ytest):
    """Function to find accuracy of classification algorithms
    Arguments:
    clf: Instance of Classifier
    Xtrain: Training set of independent features
    Xtest: Test set of independent feature
    Ytrain : Training set of dependent feature
    Ytest: Test set of independent feature
    """
    # Fit the train set to the clasifier
    clf.fit(Xtrain,Ytrain)
    # Predict output using classifier
    y_pred_clf_train = clf.predict(Xtrain)
    y_pred_clf_test = clf.predict(Xtest)
    #Find the train Accuracy
    print(str(clf.__class__).split(".")[-1].split(">")[0])
    print("Training Accuracy is {:.3f} ".format(accuracy_score(Ytrain,y_pred_clf_train)))
    print("Testing Accuracy is {:.3f} ".format(accuracy_score(Ytest,y_pred_clf_test)))
    print("----------------------------")
    print("Parameters")
    print(clf.get_params())   
```

```python
def Hyperparameter_tuning(clf,Xtrain,Xtest,Ytrain,Ytest,param):
    """Function for hyperparameter tuning for classification algorithm
    Arguments:
    ----------
    clf: Instance of classifier
    param: parameter grid for the classifier
    Xtrain: Training set of independent features
    Xtest: Test set of independent feature
    Ytrain : Training set of dependent feature
    Ytest: Test set of independent feature"""
    # Instantiate the GridSearchCV object:
    clf_cv= GridSearchCV(clf,param_grid=param,cv=5)
    # Fit it to the data
    clf_cv.fit(Xtrain,Ytrain)
    print(str(clf.__class__).split(".")[-1].split(">")[0])
    print("Best parameters are {}".format(clf_cv.best_params_))
    print("Best Score {:.3f} ".format(clf_cv.best_score_))
    return clf_cv.best_params_,clf_cv.best_score_
```

```python
def classifier_report(clf,Xtrain,Xtest,Ytrain,Ytest):
    """Function to print confusion matrix ,classifier report,roc_auc_score
    roc_auc_curve
    Arguments:
   -------------
    clf: Instance of Classifier
    Xtrain: Training set of independent features
    Xtest: Test set of independent feature
    Ytrain : Training set of dependent feature
    Ytest: Test set of independent feature"""
    try:
        #fit the data
        clf.fit(Xtrain,Ytrain)
        #Predict the label
        y_pred_clf = clf.predict(Xtest)
        print(str(clf.__class__).split(".")[-1].split(">")[0])
        print("------------------------------------------------------------------------------------")
        #Confusion Matrix
        print("Confusion Matrix")
        print("------------------------------------------------------------------------------------")
        cm=confusion_matrix(Ytest, y_pred_clf,labels=clf.classes_)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=clf.classes_)
        disp.plot()
        plt.show()
        print("------------------------------------------------------------------------------------")
        #Classification report
        print("Classification report")
        print(classification_report(Ytest, y_pred_clf))
        print("------------------------------------------------------------------------------------")
        #Roc_auc_score
        y_pred_prob_clf = clf.predict_proba(Xtest)[:,1]
        print("ROC_AUC_SCORE : {} ".format(round(roc_auc_score(Ytest,y_pred_prob_clf),3)))
        print("------------------------------------------------------------------------------------")
        #Roc_auc_curve
        # Generate ROC curve values: fpr, tpr, thresholds
        fpr,tpr,thresholds=roc_curve(Ytest,y_pred_prob_clf)
        # Plot ROC curve
        plt.plot([0,1],[0,1],'r--')
        plt.plot(fpr,tpr)
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title("ROC curve for "+str(clf.__class__).split(".")[-1].split(">")[0])
        plt.show()
    except Exception as e:
        print(e)
```
