## Feature transformations

```python
import scipy.stats as stat
import pylab 
#### to check whether feature is guassian or normal distributed
#### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
#Exponential Transformation
df['FFMC_exp']=df.FFMC**(1/1.2)
plot_data(df,'FFMC_exp')
```

```python
#logarithimic transformation
df['FFMC_log'] =np.log(df['FFMC'])
plot_data(df,'FFMC_log')
```

```python
#Boxcox transformation
df['FFMC_box'] ,parameters=stat.boxcox(df['FFMC'])
plot_data(df,'FFMC_box')
```

Box-Cox transformation of FFMC gives a dataset close to normal distribution

```python
## scaling the data while calculating vif score
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor

def vif_score(data,col):
    y = data[col]
    X = data.drop([col],axis=1)
    scalar = StandardScaler()
    X_scaled = scalar.fit_transform(X)
    vif = pd.DataFrame()
    vif['VIF'] = [variance_inflation_factor(X_scaled,i) for i in range(X_scaled.shape[1])]
    vif["Features"] = X.columns
    return vif
vif_score(df.drop(columns=['Classes','FFMC_log','FFMC_exp','FFMC_exponential',
                           'BUI','year','day','FWI','FFMC']),'Temperature')
```
